```python
def get_month_experience(work_exp:str):
    if work_exp is np.nan or work_exp is None or work_exp == ' ':
        return np.nan
    
    months_tuple = ('',)
    years_tuple = ('', '')
    work_exp_list = work_exp.split()[2:6]
    years = 0
    months = 0
    
    for i in range(len(work_exp_list) - 1):
        current_string = work_exp_list[i]
        next_string = work_exp_list[i+1]
        
        # e.g. '10 ': 10.isnumeric() -> True, '' in years_tuple -> True
        if current_string.isnumeric() and next_string[:3] in years_tuple:
            years = int(current_string)
        # e.g. '8 ': 8.isnumeric() -> True, '' in months_tuple -> True
        elif current_string.isnumeric() and next_string[:3] in months_tuple:
            months = int(current_string)
    
    return years*12 + months

data_hh['  ()'] = data_hh[' '].apply(get_month_experience)
data_hh.drop(' ', axis=1, inplace=True)
median_exp = data_hh['  ()'].median()
print(f'   ( ): {median_exp}')
```

```python
# Exchange Rates
data_currency = pd.read_csv('data/ExchangeRates.csv', sep=',')
data_currency = data_currency[['currency', 'date', 'proportion', 'close']]
data_currency['date'] = pd.to_datetime(data_currency['date'], dayfirst=True)

# CV update date -> YYYY-MM-DD
data_hh[' '] = pd.to_datetime(data_hh[' '].apply(lambda date: date.split()[0]), dayfirst=True)

# Numeric salary
data_hh_salary_numeric = data_hh[''].apply(lambda salary: salary.split()[0]).astype(dtype='int')

# ISO currency
dict_iso = {
    '.': 'UAH',
    'USD': 'USD',
    'EUR': 'EUR',
    '.': '.',
    '..': 'BYN',
    'KGS': 'KGS',
    '': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'   
}

def convert_to_iso(currency:str):
    # Structure of currency is {%currency% %currency_name%}, e.g. {29000 .}
    currency_name = currency.split()[1] 
    return dict_iso[currency_name]
        
data_hh_salary_iso = data_hh[''].apply(convert_to_iso)
data_hh_salary_iso.name = ''

# Concat salary data and CV update date
salary_data = pd.concat([data_hh[' '], data_hh_salary_numeric, data_hh_salary_iso], axis=1)

# Merge salary data with exchange rates
merged_salary_currency = salary_data.merge(
    data_currency,
    how='left',
    left_on=[' ', ''],
    right_on=['date', 'currency']
)

# Result column
salary_rub = (
    merged_salary_currency['close'].fillna(1) /
    merged_salary_currency['proportion'].fillna(1) *
    merged_salary_currency['']
)
salary_rub.name = ' ()'

# Add result column to main dataframe
data_hh = pd.concat([data_hh, salary_rub], axis=1)
data_hh.drop('', axis=1, inplace=True)
median_rub = data_hh[' ()'].median()
print(f' /  : {median_rub}')
```

```python
def outliers_z_score(data:pd.DataFrame, feature, left=3, right=4, log_scale=False, log_adder=0):
    if log_scale:
        x = np.log(data[feature] + log_adder)
    else:
        x = data[feature]
    
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left*sigma
    upper_bound = mu + right*sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned 

log_age = np.log(data_hh[''])
log_age_asymmetric = round(log_age.skew(), 3)

log_age_hist = px.histogram(
    log_age,
    x=log_age.values,
    labels={'x': ' (  )'},
    title='  ',
    width=1000
)
log_age_hist.update_layout(
    {
        'shapes': [
            {
            'type': 'line',
            'line_dash': 'dash',
            'x0': log_age.mean(),
            'y0': 3000,
            'x1': log_age.mean(),
            'y1': 0
            },
            {
            'type': 'line',
            'line_dash': 'dash',
            'x0': log_age.mean() - 3*log_age.std(),
            'y0': 3000,
            'x1': log_age.mean() - 3*log_age.std(),
            'y1': 0
            },
            {
            'type': 'line',
            'line_dash': 'dash',
            'x0': log_age.mean() + 3*log_age.std(),
            'y0': 3000,
            'x1': log_age.mean() + 3*log_age.std(),
            'y1': 0
            }
        ]
    }
)
log_age_hist.show()

outliers_age, cleaned_age = outliers_z_score(data_hh, '', log_scale=True, right=4)
data_hh = cleaned_age
display(outliers_age)
print(f'  : {log_age_asymmetric},     , \
     .')
```
